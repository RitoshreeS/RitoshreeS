```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

```

```python
df=pd.read_csv('Hiring.csv') #Read the dataset in Colab
df.head()
```

EXPLORING THE DATA





```python
df.shape  # Prints the number of rows as well as columns in a data frame
```

```python
df.info() # Index, Datatype, and Memory details

```

```python
df.value_counts(dropna=False) # Views unique values and counts
```

```python
df.fillna(df.mean()) # Replaces all null values with the mean (mean can be replaced with almost any function from the statistics section)

```

```python
sex_map = {'Male':1, 'Female':0}
df['Gender'] = df['Gender'].map(sex_map)

pyth_map = {'Yes':1, 'No':0}
df['Python_exp'] = df['Python_exp'].map(pyth_map)

edu_map = {'Graduate':1, 'Not Graduate':0}
df['Education'] = df['Education'].map(edu_map)

intern_map = {'Yes':1, 'No':0}
df['Internship'] = df['Internship'].map(intern_map)

locat_map = {'Urban':3, 'Semiurban':2, 'Rural':1}
df['Location'] = df['Location'].map(locat_map)

status_map = {'Y':1, 'N':0}
df['Recruitment_Status'] = df['Recruitment_Status'].map(status_map)


df.head()
```

```python
df['Gender'].fillna(0, inplace=True)

df['Experience_Years'].fillna(0, inplace=True)

df['Python_exp'].fillna(0, inplace=True)

df['Offer_History'].fillna(1, inplace=True)

df['Internship'].fillna(np.mean(df['Internship']), inplace=True)

df['Salary * 10E4'].fillna(np.mean(df['Salary * 10E4']), inplace=True)

df.describe()
```

```python
df.fillna(df.mean()) # Replaces all null values with the mean (mean can be replaced with almost any function from the statistics section)
```

```python
df.corr()
```

```python
m = np.random.rand(len(df))<0.72

train = df[m]
test = df[~m]
```

```python
x_train = np.asarray(train[['Gender', 'Python_exp', 'Experience_Years', 'Education', 'Internship', 'Score', 'Salary * 10E4', 'Offer_History']])
y_train = np.asarray(train[['Recruitment_Status']])

x_test = np.asarray(test[['Gender', 'Python_exp', 'Experience_Years', 'Education', 'Internship', 'Score', 'Salary * 10E4', 'Offer_History']])
y_test = np.asarray(test[['Recruitment_Status']])
```

```python
dtc = DecisionTreeClassifier(criterion="entropy", max_depth=4)
dtc.fit(x_train, y_train)
```

```python
y_test_pred = dtc.predict(x_test)
print("Expected: ",np.mean(y_test))
print("Predicted: ",np.mean(y_test_pred))
```

```python
metrics.accuracy_score(y_test, y_test_pred)
```
